11. **Idealna liczba dzieci wg płci i ogółu populacji przebadanej populacji**

W pierwszym etapie oszacowano średnią idealną liczbę dzieci dla całej populacji oraz w podziale na płeć. Obliczono również rozkład procentowy odpowiedzi według płci. Analizę przeprowadzono z wykorzystaniem zmiennych „v19” i „sex” oraz wagi analitycznej WEIGHT_COM, zapewniającej reprezentatywność wyników. Wyniki przedstawiono na wykresie słupkowym (tab. 1).



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Łączenie z dyskiem Google
from google.colab import drive
drive.mount('/content/drive')
# ŚCIEŻKA DO PLIKU
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'


df = pd.read_spss(file_path, convert_categoricals=False)
df.columns = df.columns.str.lower()

# 2. Czyszczenie danych (Waga i v19)
df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)

def clean_v19(val):
    try:
        num = float(val)
        # Wykluczamy ujemne i >= 90 (95, 98, 99)
        return num if 0 <= num < 90 else np.nan
    except:
        return np.nan

df['v19_num'] = df['v19'].apply(clean_v19)
df_valid = df.dropna(subset=['v19_num']).copy()

# 3. Obliczanie ŚREDNICH WAŻONYCH
def get_mean(data):
    return (data['v19_num'] * data['weight_com']).sum() / data['weight_com'].sum()

mean_all = get_mean(df_valid)
mean_male = get_mean(df_valid[df_valid['sex'] == 1.0])
mean_female = get_mean(df_valid[df_valid['sex'] == 2.0])

df_means = pd.DataFrame({
    'Grupa': ['Mężczyźni', 'Kobiety', 'Cały Kraj'],
    'Średnia': [mean_male, mean_female, mean_all]
})

# 4. Przygotowanie danych do rozkładu %
def group_v19(num):
    if num == 0: return '0 (Brak)'
    if num >= 6: return '6+'
    return str(int(num))

df_valid['liczba_dzieci'] = df_valid['v19_num'].apply(group_v19)

ct_pct = pd.crosstab(
    index=df_valid['liczba_dzieci'],
    columns=df_valid['sex'],
    values=df_valid['weight_com'],
    aggfunc='sum',
    normalize='columns'
) * 100

kolejnosc = ['0 (Brak)', '1', '2', '3', '4', '5', '6+']
ct_pct = ct_pct.reindex(kolejnosc).fillna(0)
ct_pct.columns = ['Mężczyźni', 'Kobiety']

plot_dist = ct_pct.reset_index().melt(id_vars='liczba_dzieci', var_name='Płeć', value_name='Procent')

# 5. WIZUALIZACJA (Dwa wykresy)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7), gridspec_kw={'width_ratios': [2, 1]})
sns.set_style("white")

# --- WYKRES 1: Rozkład % ---
colors_sex = {'Mężczyźni': '#1f77b4', 'Kobiety': '#d62728'}
sns.barplot(data=plot_dist, x='liczba_dzieci', y='Procent', hue='Płeć', palette=colors_sex, ax=ax1, edgecolor='black')
ax1.set_title('Rozkład preferencji (w %)', fontsize=15, fontweight='bold')
ax1.set_ylabel('Procent respondentów (%)')
ax1.set_xlabel('Liczba dzieci')

# Etykiety %
for p in ax1.patches:
    h = p.get_height()
    if h > 0.5:
        ax1.annotate(f'{h:.1f}%', (p.get_x() + p.get_width() / 2., h + 0.5), ha='center', fontsize=9, fontweight='bold')

# --- WYKRES 2: Porównanie Średnich ---
colors_mean = ['#1f77b4', '#d62728', '#7f7f7f'] # Niebieski, Czerwony, Szary (Kraj)
sns.barplot(data=df_means, x='Grupa', y='Średnia', palette=colors_mean, ax=ax2, edgecolor='black')
ax2.set_title('Średnia liczba dzieci', fontsize=15, fontweight='bold')
ax2.set_ylim(0, max(df_means['Średnia']) * 1.2)

# Etykiety wartości nad słupkami średniej
for i, v in enumerate(df_means['Średnia']):
    ax2.text(i, v + 0.05, f'{v:.2f}', ha='center', fontsize=12, fontweight='bold')

sns.despine()
plt.tight_layout()
plt.show()

# Raport tekstowy
print(f"Raport końcowy:")
print(f"- Średnia krajowa: {mean_all:.2f}")
print(f"- Średnia mężczyzn: {mean_male:.2f}")
print(f"- Średnia kobiet: {mean_female:.2f}")
```

13. **Płeć i idealna liczba dzieci. Czy jest istotna statystycznie?**

Dzięki wykorzystaniu regresji ważoną (WLS), można stwierdzić, że średnia idealna liczba dzieci w rodzinie dla mężczyzn wynosi 2,48, natomiast dla kobiet jest ona średnio o 0,01 niższa (dokładnie 0,013).

Współczynnik przy zmiennej płeć nie jest istotny statystycznie (p = 0,198 przy poziomie istotności α = 0,05; p \> 0,05), co oznacza, że nie ma podstaw do stwierdzenia istotnej różnicy między kobietami a mężczyznami w zakresie deklarowanej idealnej liczby dzieci.

Wartość współczynnika determinacji R² = 0,00 wskazuje, że płeć nie wyjaśnia zróżnicowania zmiennej „idealna liczba dzieci w rodzinie”.

Test F ocenia istotność modelu jako całości. Ponieważ jego wartość jest nieistotna (p = 0,198 \> 0,05), model regresji nie jest istotny statystycznie.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf


# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'

df.columns = df.columns.str.lower()

# 2. Czyszczenie danych - KLUCZOWE dla regresji
# v19: usuwamy kody specjalne (<0 i >=90)
def clean_v19(val):
    try:
        num = float(val)
        return num if 0 <= num < 90 else np.nan
    except:
        return np.nan

df['v19_clean'] = df['v19'].apply(clean_v19)

# sex: upewniamy się, że to 0 i 1 (dla łatwiejszej interpretacji)
# Jeśli 1.0 = Male, 2.0 = Female, to po odjęciu 1 mamy 0=M, 1=K
df['is_female'] = pd.to_numeric(df['sex'], errors='coerce') - 1

# waga: musi być liczbą dodatnią
df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)

# Usuwamy brakujące dane, bo regresja ich nie przyjmie
df_reg = df.dropna(subset=['v19_clean', 'is_female', 'weight_com']).copy()

# 3. MODEL WLS (Weighted Least Squares)
# Formuła: v19_clean ~ is_female
# Wagi: weight_com
model = smf.wls('v19_clean ~ is_female', data=df_reg, weights=df_reg['weight_com']).fit()

# 4. WYŚWIETLENIE WYNIKÓW
print(model.summary())

# Dodatkowa interpretacja dla Ciebie:
intercept = model.params['Intercept']
coef_sex = model.params['is_female']
p_value = model.pvalues['is_female']

print("\n" + "="*50)
print("INTERPRETACJA MODELU:")
print(f"Średnia bazowa (Mężczyźni): {intercept:.3f}")
print(f"Różnica u kobiet:           {coef_sex:+.3f}")
print(f"Istotność statystyczna (p): {p_value:.4f}")
if p_value < 0.05:
    print("Wniosek: Płeć ma STATYSTYCZNIE ISTOTNY wpływ na chęć posiadania dzieci.")
else:
    print("Wniosek: Różnica między płciami jest NIEISTOTNA statystycznie.")
print("="*50)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pycountry

# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'

# Funkcja do nazw państw (bez flag, by uniknąć błędów czcionki)
def get_country_name(iso_code):
    try:
        iso_str = str(int(float(iso_code))).zfill(3)
        country = pycountry.countries.get(numeric=iso_str)
        return country.name if country else f"Country {iso_str}"
    except:
        return f"Code {iso_code}"

# ────────────────────────────────────────────────
# Wczytanie danych (bez konwersji kategorii, żeby zachować kody)
# ────────────────────────────────────────────────
df = pd.read_spss(file_path, convert_categoricals=False)



# 1. Funkcja do nazw państw
def get_country_name(iso_code):
    try:
        iso_str = str(int(float(iso_code))).zfill(3)
        country = pycountry.countries.get(numeric=iso_str)
        return country.name if country else f"Country {iso_str}"
    except:
        return f"Code {iso_code}"

# 2. Przygotowanie danych
df['country_name'] = df['country'].apply(get_country_name)

# 3. CZYSZCZENIE DANYCH DO ŚREDNIEJ
# Zamieniamy kody specjalne na NaN (Not a Number), aby pandas je ignorował w obliczeniach
def clean_v19(val):
    try:
        num = float(val)
        if num < 0 or num >= 90:  # Odfiltrowujemy -19, 95, 98, 99 itp.
            return pd.NA
        return num
    except:
        return pd.NA

df['v19_cleaned'] = df['v19'].apply(clean_v19)

# 4. Obliczanie średniej ważonej (tylko dla poprawnych odpowiedzi)
country_means = df.groupby('country_name').apply(
    lambda x: (x['v19_cleaned'].dropna() * x.loc[x['v19_cleaned'].notna(), 'WEIGHT_COM']).sum() /
              x.loc[x['v19_cleaned'].notna(), 'WEIGHT_COM'].sum()
              if x.loc[x['v19_cleaned'].notna(), 'WEIGHT_COM'].sum() > 0 else 0,
    include_groups=False
).rename('Średnia')

# 5. Macierz procentowa (tutaj możemy zostawić kody specjalne jako osobną kategorię lub pominąć)
# Dla przejrzystości macierzy użyjemy tylko wartości 0-6+
def map_display(val):
    num = clean_v19(val)
    if pd.isna(num): return "Brak danych/95"
    if num >= 6: return "6+"
    return str(int(num))

df['v19_display'] = df['v19'].apply(map_display)

ct_country = pd.crosstab(
    index=df['country_name'],
    columns=df['v19_display'],
    values=df['WEIGHT_COM'],
    aggfunc='sum'
).fillna(0)

# Usuwamy kolumnę "Brak danych/95" z macierzy, by nie zaburzała widoku struktury chęci
if "Brak danych/95" in ct_country.columns:
    ct_country = ct_country.drop(columns=["Brak danych/95"])

ct_country_pct = ct_country.div(ct_country.sum(axis=1), axis=0) * 100
final_matrix = ct_country_pct.join(country_means)

# 6. Wizualizacja
plt.figure(figsize=(14, 10))
sns.set_theme(style="white")

sns.heatmap(
    final_matrix.sort_values('Średnia', ascending=False),
    annot=True,
    fmt=".1f",
    cmap="YlGnBu",
    cbar_kws={'label': 'Procent (%) / Średnia ważona'}
)

plt.title('Idealna liczba dzieci w rodzinie', fontsize=15)
plt.ylabel('Kraje')
plt.xlabel('Liczba dzieci')
plt.tight_layout()
```

**Kraj i idealna liczba dzieci w rodzinie. Czy zmienna jest istotna statystycznie?**

Wyniki regresji ważonej wskazują, że kraj zamieszkania jest istotnym predyktorem deklarowanej idealnej liczby dzieci (p \< 0,001). Model wyjaśnia 12,7% wariancji zmiennej zależnej (R² = 0,127), co wskazuje na umiarkowaną siłę efektu. Oznacza to, że kraje w badaniu różnią się między sobą w odróżnieniu od płci z całej populacji badania.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pycountry

# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'


df.columns = df.columns.str.lower()

# 2. Czyszczenie v19 i Wagi
def clean_v19(val):
    try:
        num = float(val)
        return num if 0 <= num < 90 else np.nan
    except:
        return np.nan

df['v19_clean'] = df['v19'].apply(clean_v19)
df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)

# 3. Nazwy krajów (dla czytelności wyników)
def get_country_name(iso_code):
    try:
        iso_str = str(int(float(iso_code))).zfill(3)
        return pycountry.countries.get(numeric=iso_str).name
    except:
        return f"Country_{iso_code}"

df['country_name'] = df['country'].apply(get_country_name)

# Usuwamy braki (regresja wymaga pełnych danych w analizowanych kolumnach)
df_reg = df.dropna(subset=['v19_clean', 'country_name', 'weight_com']).copy()

# 4. MODEL WLS
# Używamy C(country_name), aby potraktować kraj jako zmienną kategoryczną
model = smf.wls('v19_clean ~ C(country_name)', data=df_reg, weights=df_reg['weight_com']).fit()

# 5. TEST GLOBALNY (czy zmienna country jako całość jest istotna)
# Wykonujemy analizę wariancji (ANOVA) na modelu WLS
anova_table = sm.stats.anova_lm(model, typ=2)
p_value_global = anova_table.loc['C(country_name)', 'PR(>F)']

# 6. RANKING KRAJÓW (na podstawie przewidywanych średnich z modelu)
# Wyciągamy średnie dla każdego kraju bezpośrednio z predykcji modelu
countries = df_reg['country_name'].unique()
country_means = []

for c in countries:
    # Obliczamy średnią ważoną dla każdego kraju
    subset = df_reg[df_reg['country_name'] == c]
    w_mean = (subset['v19_clean'] * subset['weight_com']).sum() / subset['weight_com'].sum()
    country_means.append({'Kraj': c, 'Średnia': w_mean})

ranking = pd.DataFrame(country_means).sort_values(by='Średnia', ascending=False)

# 7. WYNIKI
print("\n" + "="*50)
print("WYNIKI REGRESJI WAŻONEJ (WLS) DLA PAŃSTW")
print("="*50)
print(f"P-value (Test globalny): {p_value_global:.4e}")
print(f"R-squared (R²):         {model.rsquared:.4f}")
print("-" * 50)

if p_value_global < 0.05:
    print("WNIOSEK: Kraj zamieszkania ma STATYSTYCZNIE ISTOTNY wpływ na deklaracje.")
else:
    print("WNIOSEK: Brak istotnych różnic między krajami.")

print("\nTOP 3 KRAJE (Najwyższa chęć posiadania dzieci):")
print(ranking.head(3).to_string(index=False))

print("\nBOTTOM 3 KRAJE (Najniższa chęć posiadania dzieci):")
print(ranking.tail(3).to_string(index=False))
print("="*50)
```

15. **Idealna liczba dzieci w rodzinie w podziale na wiek**

Następnie przeanalizowano idealną liczbę dzieci w podziale na grupy wiekowe. W celu zwiększenia czytelności respondentów podzielono na następujące kategorie: 18–24, 25–34, 35–44, 45–54, 55–64 oraz 65+. Wyniki przedstawiono w formie mapy cieplnej (tab. 3).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'

df = pd.read_spss(file_path, convert_categoricals=False)

# --- NAPRAWA: Standaryzacja nazw kolumn ---
df.columns = df.columns.str.lower()
# Teraz 'AGE' stanie się 'age', a 'SEX' stanie się 'sex' itd.
# ------------------------------------------

# 2. CZYSZCZENIE DANYCH
df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)

def clean_v19(val):
    try:
        num = float(val)
        if num < 0 or num >= 90:
            return np.nan
        return num
    except:
        return np.nan

df['v19_cleaned'] = df['v19'].apply(clean_v19)

# 3. CZYSZCZENIE I GRUPOWANIE WIEKU
# Używamy już małych liter dzięki krokowi powyżej
df['age_num'] = pd.to_numeric(df['age'], errors='coerce')
df = df[df['age_num'] > 0].copy()

bins = [0, 24, 34, 44, 54, 64, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['age_num'], bins=bins, labels=labels)

# 4. OBLICZANIE ŚREDNIEJ WAŻONEJ
def get_weighted_mean(group):
    valid = group.dropna(subset=['v19_cleaned'])
    if len(valid) == 0: return 0
    # Używamy weight_com (małymi literami)
    return (valid['v19_cleaned'] * valid['weight_com']).sum() / valid['weight_com'].sum()

age_means = df.groupby('age_group', observed=True).apply(
    get_weighted_mean,
    include_groups=False
).rename('Średnia')

# 5. TWORZENIE MACIERZY PROCENTOWEJ
df['v19_display'] = df['v19_cleaned'].apply(
    lambda x: '6+' if x >= 6 else (str(int(x)) if pd.notna(x) else np.nan)
)

ct_age = pd.crosstab(
    index=df['age_group'],
    columns=df['v19_display'],
    values=df['weight_com'],
    aggfunc='sum'
).fillna(0)

ct_age_pct = ct_age.div(ct_age.sum(axis=1), axis=0) * 100
final_age_matrix = ct_age_pct.join(age_means)

# 6. WIZUALIZACJA
plt.figure(figsize=(12, 8))
sns.set_theme(style="white")

sns.heatmap(
    final_age_matrix,
    annot=True,
    fmt=".1f",
    cmap="YlOrRd",
    cbar_kws={'label': 'Procent (%) / Średnia'}
)

plt.title('Idealna liczba dzieci wg grup wiekowych', fontsize=15)
plt.ylabel('Grupa wiekowa')
plt.xlabel('Liczba dzieci i średnia')
plt.tight_layout()
plt.show()
```

Różnice między grupami wiekowymi są minimalne, jednak młodsze osoby częściej wskazują skrajne odpowiedzi (0–1 oraz 3–6+), podczas gdy w starszych grupach odpowiedzi są bardziej skoncentrowane wokół średniej.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.stats.weightstats import DescrStatsW


# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'


# 2. NATYCHMIASTOWA standaryzacja nazw (to rozwiązuje KeyError)
df.columns = df.columns.str.lower()

# Teraz nazwy to na 100% 'weight_com', 'v19', 'age', 'sex'
# 3. Czyszczenie wagi
if 'weight_com' in df.columns:
    df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)
else:
    # Failsafe: jeśli waga nazywa się inaczej, stwórzmy kolumnę z 1.0
    df['weight_com'] = 1.0

# 4. Czyszczenie zmiennej v19
def clean_v19(val):
    try:
        num = float(val)
        # Odfiltrowujemy kody specjalne (ujemne i >= 90)
        return num if 0 <= num < 90 else np.nan
    except:
        return np.nan

df['v19_cleaned'] = df['v19'].apply(clean_v19)

# 5. Czyszczenie i grupowanie wieku
df['age_num'] = pd.to_numeric(df['age'], errors='coerce')
df_clean = df[(df['age_num'] > 0) & (df['v19_cleaned'].notna())].copy()

bins = [0, 24, 34, 44, 54, 64, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df_clean['age_group'] = pd.cut(df_clean['age_num'], bins=bins, labels=labels)

# 6. Funkcja statystyk ważonych
def get_weighted_stats(group):
    v = group['v19_cleaned'].values
    w = group['weight_com'].values

    if len(v) == 0 or w.sum() <= 0:
        return pd.Series([0, 0, 0, 0, 0], index=['min', 'q1', 'med', 'q3', 'max'])

    d = DescrStatsW(v, weights=w)
    # Obliczamy kwantyle (0, 0.25, 0.5, 0.75, 1)
    q = d.quantile([0, 0.25, 0.5, 0.75, 1]).values
    return pd.Series(q, index=['min', 'q1', 'med', 'q3', 'max'])

# 7. Obliczanie statystyk z wyciszeniem ostrzeżenia o include_groups
stats = df_clean.groupby('age_group', observed=True).apply(
    get_weighted_stats,
    include_groups=False
)

# Używamy próbkowania ważonego (10k rekordów), by pokazać kształt rozkładu bez błędu pamięci
df_sample = df_clean.sample(n=10000, replace=True, weights=df_clean['weight_com'])

plt.figure(figsize=(12, 7))
sns.violinplot(
    data=df_sample,
    x='age_group',
    y='v19_cleaned',
    hue='age_group',
    palette='muted',
    inner='quartile', # pokaże linie Q1, Med i Q3 wewnątrz "skrzypiec"
    cut=0 # nie dorysowuje sztucznych ogonów poza zakres danych
)

plt.title('Gęstość deklarowanej idealnej liczby dzieci w rodzinie wg wieku (Próba ważona)', fontsize=14)
plt.ylabel('Liczba dzieci')
plt.ylim(-1, 6) #
```

17. **Wiek i idealna liczba dzieci. Czy jest istotna statystycznie?**

Wiek ma formalnie istotny wpływ na zmienną, gdyż p-value wynosi 0.0218 \< 0.05, **ale praktycznie znikomy wpływ** na deklarowaną idealną liczbę dzieci. Na każde 100 lat życia (hipotetycznie) efekt wynosiłby ok. 0.0007 dziecka, co pokazuje, że w praktyce wiek jest bardzo słabym predyktorem.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf


# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'

df.columns = df.columns.str.lower()

# 2. Czyszczenie danych
def clean_v19(val):
    try:
        num = float(val)
        return num if 0 <= num < 90 else np.nan
    except:
        return np.nan

df['v19_clean'] = df['v19'].apply(clean_v19)
df['age_num'] = pd.to_numeric(df['age'], errors='coerce')
df['weight_com'] = pd.to_numeric(df['weight_com'], errors='coerce').fillna(1)

# Usuwamy braki danych i wiek <= 0 (kody błędów)
df_reg = df.dropna(subset=['v19_clean', 'age_num', 'weight_com']).copy()
df_reg = df_reg[df_reg['age_num'] > 0]

# 3. Model WLS (Regresja liniowa ważona)
# Formuła: v19_clean ~ age_num
model = smf.wls('v19_clean ~ age_num', data=df_reg, weights=df_reg['weight_com']).fit()

# 4. Wyciągnięcie wyników
results = pd.concat([
    model.params,
    model.pvalues,
    model.conf_int()],
    axis=1
)
results.columns = ['Współczynnik (Coef)', 'p-value', '[0.025', '0.975]']

# 5. Wyświetlenie raportu
print("\n" + "="*60)
print("WYNIKI REGRESJI WLS: WPŁYW WIEKU NA V19")
print("="*60)
print(f"R-squared: {model.rsquared:.4f}")
print("-" * 60)
print(results.round(4).to_string())
print("="*60)

# Dodatkowa interpretacja
coef_age = model.params['age_num']
p_age = model.pvalues['age_num']
print(f"\nInterpretacja: Z każdym rokiem życia deklarowana idealna liczba dzieci")
print(f"{'rośnie' if coef_age > 0 else 'spada'} średnio o {abs(coef_age):.4f}.")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pycountry


# ────────────────────────────────────────────────
# Łączenie z dyskiem Google
# ────────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')
# ────────────────────────────────────────────────
# ŚCIEŻKA DO PLIKU
# ────────────────────────────────────────────────
file_path = '/content/drive/MyDrive/ZA10000_v2-0-0(1).sav'
df = pd.read_spss(file_path, convert_categoricals=False)

# Czyszczenie v19, wieku i wagi
df['v19_clean'] = df['v19'].apply(lambda x: float(x) if 0 <= float(x) < 90 else np.nan)
df['AGE_num'] = pd.to_numeric(df['AGE'], errors='coerce')
df['WEIGHT_COM'] = pd.to_numeric(df['WEIGHT_COM'], errors='coerce').fillna(1)

# Płeć jako zmienna binarna (0 = Mężczyzna, 1 = Kobieta)
df['is_female'] = pd.to_numeric(df['SEX'], errors='coerce') - 1

# Nazwy krajów
def get_country_name(iso_code):
    try:
        iso_str = str(int(float(iso_code))).zfill(3)
        return pycountry.countries.get(numeric=iso_str).name
    except: return f"country_{iso_code}"
df['country_name'] = df['country'].apply(get_country_name)

# Filtracja braków
df_model = df.dropna(subset=['v19_clean', 'is_female', 'AGE_num', 'country_name']).copy()
df_model = df_model[df_model['AGE_num'] > 0]

# 2. MODEL REGRESJI WIELOZMIENNEJ (WLS)
# v19_clean = b0 + b1*is_female + b2*AGE_num + b3*country
formula = 'v19_clean ~ is_female + AGE_num + C(country_name)'
model = smf.wls(formula, data=df_model, weights=df_model['WEIGHT_COM']).fit()

# 3. ANALIZA ISTOTNOŚCI (Testy typu II dla zmiennych kategorycznych)
anova_table = sm.stats.anova_lm(model, typ=2)

# 4. RAPORT
print("\n" + "="*60)
print("PEŁNY RAPORT MODELU WIELOZMIENNEGO (WLS)")
print("="*60)
print(f"R-squared (R²): {model.rsquared:.4f}")
print("-" * 60)

# Czy płeć istotna?
p_SEX = model.pvalues['is_female']
coef_SEX = model.params['is_female']
print(f"PŁEĆ:  p-value = {p_SEX:.4f} | Coef = {coef_SEX:+.3f}")
print(f"       -> {'Istotna' if p_SEX < 0.05 else 'Nieistotna'} po kontroli wieku i kraju.")

# Czy wiek istotny?
p_AGE = model.pvalues['AGE_num']
coef_AGE = model.params['AGE_num']
print(f"WIEK:  p-value = {p_AGE:.4f} | Coef = {coef_AGE:+.4f}")
print(f"       -> {'Istotny' if p_AGE < 0.05 else 'Nieistotny'}.")

# Czy kraj nadal silny? (Test globalny)
p_country = anova_table.loc['C(country_name)', 'PR(>F)']
print(f"KRAJ:  p-value = {p_country:.4e}")
print(f"       -> {'Nadal wykazuje bardzo silny wpływ' if p_country < 0.05 else 'Nieistotny'}.")
print("="*60)

# Ranking 5 najsilniejszych efektów krajowych (względem bazy)
print("\nTop 5 współczynników krajowych (odchylenie od bazy):")
print(model.params.filter(like='country_name').sort_values(ascending=False).head(5))
```
